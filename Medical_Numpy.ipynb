```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Cleaned_medical_data")
data = df.to_numpy()
```

```python
age = data[:, df.columns.get_loc("Age")]
bmi = data[:, df.columns.get_loc("BMI")]
chol = data[:, df.columns.get_loc("Cholesterol")]
```

```python
print("Age mean:", np.mean(age))
print("BMI max:", np.max(bmi))
print("Cholesterol median:", np.median(chol))
```
