```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Cleaned_medical_data", index_col=0)
```

```python
df.head()
```

```python
#Distribution of Age

plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=10, kde=True)
plt.title('Age Distribution of Patients')
plt.show()
```

```python
# Count of Patients by Gender

plt.figure(figsize=(6,4))
plt.title('Number of Patients by Gender')
sns.countplot(
    data=df,
    x='Gender',
    hue='Gender',
    palette='Set2',
    legend=False
)
plt.show()
```

```python
# Boxplot of BMI by Gender
plt.figure(figsize=(5,7))
sns.boxplot(data=df, x='Gender', y='BMI',hue='Gender',legend=False,palette='pastel')
plt.title('BMI Distribution by Gender')
plt.show()
```

```python
# Scatter Plot: Cholesterol vs Blood Pressure

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Blood_Pressure', y='Cholesterol', hue='Smoker',legend=False, palette='Set2')
plt.title('Cholesterol vs Blood Pressure by Smoker Status')
plt.show()
```

```python
# Violin Plot: BMI by Diagnosis

plt.figure(figsize=(5,7))
sns.violinplot(data=df,x='Diagnosis', y='BMI',hue='Diagnosis',legend=False, palette='muted')
plt.title("BMI Distribution by Diagnosis")
```
